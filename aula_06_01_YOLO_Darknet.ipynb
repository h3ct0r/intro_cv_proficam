```
# Testando o detector com o darknet

!./darknet detect cfg/yolov4.cfg yolov4.weights data/person.jpg
```

```
# Função para visualizar o resultado da detecção com opencv e matplotlib

def mostrar(caminho):
  imagem = cv2.imread(caminho)
  fig = plt.gcf()
  fig.set_size_inches(18, 10)
  plt.axis('off')
  plt.imshow(cv2.cvtColor(imagem, cv2.COLOR_BGR2RGB))
  plt.show()
```

```
# Visualizando o resultado

mostrar('predictions.jpg')
```

```
# Detecta e mostra objetos de outra imagem

!./darknet detect cfg/yolov4.cfg yolov4.weights data/giraffe.jpg

mostrar('predictions.jpg')
```
